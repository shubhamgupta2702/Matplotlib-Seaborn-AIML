```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,12)
y = x**2
x
```

```python
plt.subplot(2,2,1)
plt.plot(x,x)
plt.subplot(2,2,4)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(y,x)
plt.subplot(2,2,3)
plt.plot(x,x**2)
```

```python
fig = plt.figure()
axis1 = fig.add_axes([0.1,0.1,0.7,1])  #left,bottom,width,height.
axis1.plot(x,y)
axis2 = fig.add_axes([0.3,0.5,0.3,0.3])   #left -> 30% bottom -> 50%
axis2.plot(x,y)
```

```python
fig = plt.figure(figsize=(10,4))  #figsize = (width,height), dpi-dots per inch
axis1 = fig.add_axes([0.1,0.1,0.7,1])  #left,bottom,width,height.
axis1.plot(x,y)
axis2 = fig.add_axes([0.3,0.5,0.3,0.3])   #left -> 30% bottom -> 50%
axis2.plot(x,y)
```

**Types of Plots**

```python
plt.scatter(x,y)
```

```python
plt.hist(x)
```

```python
plt.boxplot(y)
plt.savefig('boxplot.png')
```
