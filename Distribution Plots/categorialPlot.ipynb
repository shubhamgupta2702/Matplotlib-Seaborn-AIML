**Categorial Data Plots**

now lets discuss using seaborn to plot categorial data. there are a few main plot types for this:

**1.boxplot**
**2.violinplot**
**3.stripplot**
**4.swarmplot**
**5.barplot**
**6.countplot**

discussing examples of each...

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('tips')
df
```

```python
sns.countplot(df['sex'])
```

```python
sns.countplot(x =df['sex'], hue=df['smoker'])
```

```python
sns.countplot(x = df['day'],hue=df['sex'])
```

```python
sns.countplot(df['time'])
```

```python
sns.countplot(df['total_bill'])
```

```python
# x-axis have categorical data
# y-axis have quantitative data
sns.barplot(x = df['sex'], y = df['tip'])
```

**you can change the estimator object to your own function, that converts a vector to a scalar Mean(default):**
np.mean
np.median
np.sum
np.std
len
np.min
np.max

```python
import numpy as np
sns.barplot(x = df['sex'], y=df['total_bill'], estimator=np.sum)
```

```python
sns.barplot(x = df['sex'], y=df['tip'], estimator=np.sum)
```

```python
# x-axis contains quantitative data.
#y-axis contains qualitative data or categorical data.
sns.boxplot(x = 'tip', y = 'day',data=df, palette='rainbow')
```

```python
sns.violinplot(x='day', y='total_bill', data=df)
# it shows distribution - area spread shown and also data with more data.
```

**stripplot and swarmplot**


```python
sns.stripplot(x = 'day', y = 'total_bill', data=df)
```

```python
sns.swarmplot(x = 'day', y = 'total_bill', data=df)
```

```python
sns.violinplot(x = 'tip', y = 'day', data=df)
sns.swarmplot(x = 'tip', y = 'day', data=df)
```
